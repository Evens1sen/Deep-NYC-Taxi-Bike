```python
import pandas as pd
import numpy as np
# from Param import *
from datetime import datetime

TIME_START = 20190101000000  # %Y%m
TIME_END = 20201231235959  # %Y%m
TIME_STEP = 30  # minutes
```

```python
print(str(TIME_START)[0:6])
print(TIME_END)
pd.to_datetime(TIME_START)
'{}-{}-{}-in_flow.h5'.format(str(TIME_START)[0:6],str(TIME_END)[0:6],TIME_STEP)
```

```python
filepath_list = []
for i in range(1, 13):
    filepath_list.append('../../../data/mhy/NYCBikeData/npyData/new_graph/2019{}-citibike-tripdata-graph-inoutflow-{}.npz'.format(str(i).zfill(2),TIME_STEP))
for i in range(1, 13):
    filepath_list.append('../../../data/mhy/NYCBikeData/npyData/new_graph/2020{}-citibike-tripdata-graph-inoutflow-{}.npz'.format(str(i).zfill(2),TIME_STEP))
```

```python
out_flow = pd.DataFrame({'time':pd.date_range(str(TIME_START), str(TIME_END), freq='30min')})
in_flow = pd.DataFrame({'time':pd.date_range(str(TIME_START), str(TIME_END), freq='30min')})
for i in range(69):
    in_flow[str(i)] = 0
    out_flow[str(i)] = 0
days = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31,
       31, 29, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31,]
cur = 0
out_flow.iloc[[i for i in range(10)],[i for i in range(1,70)]]
```

```python
cur = 0
for i in range(len(filepath_list)):
    data = np.load(filepath_list[i])['arr_0']
    in_flow.iloc[[i for i in range(cur, cur+days[i]*24*2 )],[i for i in range(1,70)]] = pd.DataFrame(data[0,0:days[i]*24*2,0:69] + 1)
    out_flow.iloc[[i for i in range(cur, cur+days[i]*24*2 )],[i for i in range(1,70)]] = pd.DataFrame(data[0,0:days[i]*24*2,0:69] + 1)
    cur = cur +days[i]*24*2  
```
